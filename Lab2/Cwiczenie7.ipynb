# Ćwiczenie 7. Działanie wyrównania histogramu na obrazach

Biblioteki używane dla wykonania tego ćwiczenia:

```python
%matplotlib inline

import numpy as np  
import matplotlib.pyplot as plt
from PIL import Image
import cv2 as cv
```

## Wyjaśnienie teoretyczne

Wyrównywanie histogramu (ang. histogram equalization) to technika poprawy kontrastu obrazu, która polega na redystrybucji poziomów jasności (pikseli) tak, aby uzyskać bardziej równomierne rozłożenie wartości w całym zakresie tonalnym (0–255 dla obrazów w skali szarości). W naszym zadaniu wykorzystano funkcję equalizeHist z biblioteki OpenCV do automatycznego wykonania wyrównania histogramu dla obrazów testowych. Następnie porównano histogramy przed i po przekształceniu, aby zobaczyć efekt działania metody.

## 1) chest_xray.tif

```python
fileName = 'ObrazyTestowe/chest-xray.tif'  # <-- poprawка rozszerzenia

# Odczyt i konwersja na skale szarości
chestImage = np.asarray(Image.open(fileName).convert('L'))

plt.figure(figsize=[15, 10])

# Oryginalny obraz
plt.subplot(2, 2, 1)
plt.title('chest_xray - za ciemny obrazek')
plt.imshow(chestImage, cmap='gray', vmin=0, vmax=255)

# Histogram oryginalnego obrazu
plt.subplot(2, 2, 2)
plt.title('Histogram (przed)')
plt.hist(chestImage.ravel(), bins=256, range=[0, 256])

# Wyrównanie histogramu
correctedChestImage = cv.equalizeHist(chestImage)

# Poprawiony obraz
plt.subplot(2, 2, 3)
plt.title('chest_xray - po wyrównaniu')
plt.imshow(correctedChestImage, cmap='gray', vmin=0, vmax=255)

# Histogram poprawionego obrazu
plt.subplot(2, 2, 4)
plt.title('Histogram (po)')
plt.hist(correctedChestImage.ravel(), bins=256, range=[0, 256])

plt.show()
```
